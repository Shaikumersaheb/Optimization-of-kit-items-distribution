```python
from feature_engine.outliers import Winsorizer
```

```python
## *** Instantiate the Winsorizer object *** ##
winsor_iqr = Winsorizer(capping_method = "iqr",
                        tail = "both", fold = 1.5,
                        variables = ["No_of_Kits"])
```

```python
## *** Fit and transform the data *** ##
updated_kit = winsor_iqr.fit_transform(kit[["No_of_Kits"]])
```

```python
## *** Let's see boxplot *** ##
sns.boxplot(updated_kit)
kit["No_of_Kits"] = updated_kit
```

```python
sns.boxplot(kit["customer_code"])
```

```python
variance = updated_kit.var()
```

```python
near_zero_var_features = variance[variance < 0.01]
print(near_zero_var_features)
```

```python
updated_kit["No_of_Kits_bins"] = pd.cut(updated_kit["No_of_Kits"],
                                        bins = 3,
                                        labels = ["Low", "Medium", "High"])
```
